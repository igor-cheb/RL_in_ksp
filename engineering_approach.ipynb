```python
##### import numpy as np
import pandas as pd
import gym
import gym.spaces
import rocket_lander_gym

VEL_STATE = False
env = gym.make('RocketLander-v0')
env.reset()

PRINT_DEBUG_MSG = True
observation = [0]*7
angle_speed = 0
proj_angle_25 = 0
proj_angle_10 = 0
observations = []


for i in range(1000):
    env.render()
    angle_lim = 1.2
    vel_lim = 1.7
    
    if (observation[2]<angle_lim) & (observation[2]>0):
        if observation[-1]>0:
            #слева от нуля, вращение влево | отр. пшикалка
            action = [0,5, -observation[2]*observation[-1]*3]
        else:
            #слева от нуля, вращение вправо | пол. пшикалка
            action = [0,5, -observation[2]*observation[-1]*10]
    elif (observation[2]>-angle_lim) & (observation[2]<0):
        if observation[-1]>0:
            #справа от нуля, вращение влево | отр. пшикалка
            action = [0,5, observation[2]*observation[-1]*10]
        else:
            #справа от нуля, вращение вправо | положительная пшикалка
            action = [0,5, observation[2]*observation[-1]*3]
    else:
        action = [0, 3, -observation[2]*10]

    observation,reward,done,info =env.step(action)
    if PRINT_DEBUG_MSG:
        print("Action Taken  ",action)
        print("Observation   ",observation)
        print("Reward Gained ",reward)
        print("Info          ",info,end='\n\n')

    if done:
        print("Simulation done.")
        break
env.close()
```
